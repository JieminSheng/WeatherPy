```python
# 1. Read the WeatherPy_vacation.csv into a DataFrame.
vacation_df = pd.read_csv("../Vacation_Search/WeatherPy_Vacation.csv")
vacation_df.head(10)
```

```python
# 2. Using the template add the city name, the country code, the weather description and maximum temperature for the city.
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""

# 3a. Get the data from each row and add it to the formatting template and store the data in a list.
hotel_info = [info_box_template.format(**row) for index, row in vacation_df.iterrows()]

# 3b. Get the latitude and longitude from each row and store in a new DataFrame.
locations = vacation_df[["Lat", "Lng"]]
```

```python
# 4a. Add a marker layer for each city to the map.
fig = gmaps.figure()
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig = gmaps.figure(center=(50.0, -50.0), zoom_level=3.0)
fig.add_layer(marker_layer)

# 4b. Display the figure
fig
```

```python
# Available cities by countries excluding those less than four cities
number_of_cities_per_country_df = vacation_df.groupby(["Country"]).count()["City"]

number_of_cities_per_country_df = pd.DataFrame({
    "Number of Cities": number_of_cities_per_country_df})

number_of_cities_per_country_df = number_of_cities_per_country_df[(number_of_cities_per_country_df["Number of Cities"] >= 4)]

number_of_cities_per_country_df
```

```python
# Prompt the user to pick a country
selected_country = str(input("What country would you like to go to? "))
```

```python
# Filter the city_data_df DataFrame using the input statements to create a new DataFrame using the loc method.
preferred_cities_of_country_df = vacation_df.loc[(vacation_df["Country"] == selected_country)]
preferred_cities_of_country_df
```

```python
# Prompt the user to pick four cities, please note that not all cities might have routes between them
first_last_city = str(input("Which city would you like to start and end in? "))
second_city = str(input("Which city would you like to go second? "))
third_city = str(input("Which city would you like to go third? "))
fourth_city = str(input("Which city would you like to go fourth? "))
```

```python
# From the map above pick 4 cities and create a vacation itinerary route to travel between the four cities. 
# 5. Create DataFrames for each city by filtering the 'vacation_df' using the loc method. 
# Hint: The starting and ending city should be the same city.
vacation_start = preferred_cities_of_country_df.loc[preferred_cities_of_country_df["City"] ==  first_last_city]
vacation_end = preferred_cities_of_country_df.loc[preferred_cities_of_country_df["City"] == first_last_city]
vacation_stop1 = preferred_cities_of_country_df.loc[preferred_cities_of_country_df["City"] == second_city]
vacation_stop2 = preferred_cities_of_country_df.loc[preferred_cities_of_country_df["City"] == third_city] 
vacation_stop3 = preferred_cities_of_country_df.loc[preferred_cities_of_country_df["City"] == fourth_city]
```

```python
# 6. Get the latitude-longitude pairs as tuples from each city DataFrame using the to_numpy function and list indexing.
start = vacation_start["Lat"].to_numpy()[0], vacation_start["Lng"].to_numpy()[0]
end = vacation_end["Lat"].to_numpy()[0], vacation_end["Lng"].to_numpy()[0]
stop1 = vacation_stop1["Lat"].to_numpy()[0], vacation_stop1["Lng"].to_numpy()[0]
stop2 = vacation_stop2["Lat"].to_numpy()[0], vacation_stop2["Lng"].to_numpy()[0]
stop3 = vacation_stop2["Lat"].to_numpy()[0], vacation_stop3["Lng"].to_numpy()[0]
```

```python
# Prompt the user to pick a method of travel
selected_travel_mode = str(input("How would you like to travel; please select either DRIVING, BICYCLING, or WALKING. "))
```

```python
# 7. Create a direction layer map using the start and end latitude-longitude pairs,
# and stop1, stop2, and stop3 as the waypoints. The travel_mode should be "DRIVING", "BICYCLING", or "WALKING"
fig2 = gmaps.figure()
vacation_itinerary = gmaps.directions_layer(
        start, end, waypoints=[stop1, stop2, stop3],
        travel_mode= selected_travel_mode)
fig2.add_layer(vacation_itinerary)
fig2
```

```python
# 8. To create a marker layer map between the four cities.
#  Combine the four city DataFrames into one DataFrame using the concat() function.
frames = [vacation_start, vacation_stop1, vacation_stop2, vacation_stop3]


itinerary_df = pd.concat(frames,ignore_index=True)
itinerary_df
```

```python
# 9 Using the template add city name, the country code, the weather description and maximum temperature for the city. 
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""

# 10a Get the data from each row and add it to the formatting template and store the data in a list.
selected_hotel_info = [info_box_template.format(**row) for index, row in itinerary_df.iterrows()]

# 10b. Get the latitude and longitude from each row and store in a new DataFrame.
selected_locations = itinerary_df[["Lat", "Lng"]]
```

```python
# 11a. Add a marker layer for each city to the map.
fig3 = gmaps.figure()
selected_marker_layer = gmaps.marker_layer(selected_locations, info_box_content=selected_hotel_info)
fig3.add_layer(selected_marker_layer)

# 11b. Display the figure
fig3
```
